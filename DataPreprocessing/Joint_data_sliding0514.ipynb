```python
import json
import glob
import pandas as pd
import pickle
```

```python
# JSON file 경로
file_list = glob.glob('./joint_Json/*')
```

```python
# joint 값 전처리 함수
def prepro(file_list=None):
    result = [] 
    for file_path in file_list:
        with open(file_path, "r") as json_file:
            # joint data 추출부
            jsonObject = json.load(json_file)
            jsonArray = jsonObject.get('people') # joint 값 추출
            lst = []
            keypoints = []
            for list in jsonArray:
                keypoint = list.get('pose_keypoints_2d')
                for i in range(len(keypoint)): # x,y 좌표 뒤에 위치한 확률값 제외하고 가져오기
                    if i % 3 != 2:
                        lst.append(keypoint[i])
                    else:
                        pass
                keypoints.append(lst[:36])
                lst=[]
            result.append(keypoints)                            
    return result
```

```python
def CreIndexList(file_list):    
    index_list=[]
    for file_path in file_list:
        # 인덱스 list 생성
        temp=file_path.split('_')
        temp=temp[1]+temp[2]
        index=temp.replace('Json\\','')
        index_list.append(index)
    return index_list
```

```python
CreIndexList(file_list)
```

```python
# Label값 생성 함수
def Labeling(file_list=None):
    y_label=[]
    # assult=1, normal=2, smoke=3
    for file_name in file_list:
        if('assault' in file_name)==True:
            action_num=1
        elif('normal' in file_name)==True:
            action_num=2
        else:
            action_num=3
        y_label.append(action_num)
    return y_label
```

```python
result = prepro(file_list)
lst=Labeling(file_list)
index=CreIndexList(file_list)
```

```python
# 추출 Data 개수 원본 데이터 개수와 일치 여부
print(len(file_list)==len(lst)==len(result))
print(len(file_list))
```

### 각 video 마다 32개씩 밀면서 데이터 합치기

```python
#최종 dataframe 초기값
df_all_video = pd.DataFrame(index=range(0), columns=["p_1", "p_2"])

for i in range(len(index_list)):
    print("ind : ", index_list[i])
    # index)list를 순차적으로 접근
    ind = index_list[i]
    df_1_video = re_df_1.loc[ind, ["p_1", "p_2"]]
    
    # 각 비디오마다 32개씩 묶으면서 합쳐준다.
    df_sub = df_1_video[:32]
    for i in range(len(df_1_video) - 32):
        df_sub = pd.concat([df_sub,df_1_video[i+1:i + 33]])
    
    print("vieo_len :", len(df_1_video), "  sum :", len(df_sub))
    df_all_video = pd.concat([df_all_video, df_sub])
    print("df_all_video 크기 : ", len(df_all_video))
    print("")
df_all_video
```
