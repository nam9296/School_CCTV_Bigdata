## people 수 확인

```python
# JSON file 경로
file_list = glob.glob('./joint_Json/*')
```

```python
# file별 people 수 확인
for file_path in file_list:
    with open(file_path, "r") as json_file:
        json_data = json.load(json_file)
        if (len(json_data['people']) == 1) | (len(json_data['people']) > 2):
            print(json_file.name)

print(len(json_file.name))
```

## 관절 개수 확인

```python
joint = pd.read_csv('./sliding.csv')
joint
```

```python
# joint 개수 확인을 위해 결측치 제거
joint.isnull().sum()
joint_drop = joint.dropna()
```

```python
# 사람별 joint 개수 확인
p1_cnt = []
p2_cnt = []

for i in joint_drop['p_1']:
    p1_cnt.append(len(re.split(',',i))!=36)

for i in joint_drop['p_2']:
    p2_cnt.append(len(re.split(',',i))!=36)
    
print(p1_cnt=='True')
print(p2_cnt=='True')
```
