```python
import json
import glob
```

```python
file_list = glob.glob('./joint_Json/*')
```

```python
def prepro(file_list):
    result = [] 
    for file_path in file_list:
        with open(file_path, "r") as json_file:
            jsonObject = json.load(json_file)
            jsonArray = jsonObject.get('people')
            lst = []
            keypoints = []
            for list in jsonArray:
                keypoint = list.get('pose_keypoints_2d')
                for i in range(len(keypoint)): # 확률값 제외하고 가져오기
                    if i % 3 != 2:
                        lst.append(keypoint[i])
                    else:
                        pass
                keypoints.append(lst[:36])
                lst=[]
            result.append(keypoints)
    return result
```

```python
result = prepro(file_list)
```
