```python
import json
import pandas as pd
import pickle
import json
import glob, os
```

```python
file_list = glob.glob('./joint_Json/*')
```

```python
# joint 값 전처리 함수
def prepro(file_list=None):
    result = [] 
    for file_path in file_list:
        with open(file_path, "r") as json_file:
            # 인덱스로 사용할 이름 split
            temp=file_path.split('_')
            index_name = temp[1]+temp[2]
            
            # joint data 추출부
            jsonObject = json.load(json_file)
            jsonArray = jsonObject.get('people') # joint 값 추출
            lst = []
            keypoints = []
            
            for list in jsonArray:
                keypoint = list.get('pose_keypoints_2d')
                for i in range(len(keypoint)): # x,y 좌표 뒤에 위치한 확률값 제외하고 가져오기
                    if i % 3 != 2:
                        lst.append(keypoint[i])
                    else:
                        pass
                keypoints.append(index_name)
                keypoints.append(lst[:36])
                lst=[]
            result.append(keypoints)   
            
    return result
```

```python
pre = prepro(file_list)
```

```python
pre
```

```python
with open('./output_result1.txt', 'w') as text_file:
    for i in range(len(pre)): # 전체 
        for j in range(len(pre[i])): # 사람1, 2
            for k in range(36): # joint 값
                text_file.write('{}'.format(pre[i][j][k]))
                if j < 35:
                    text_file.write(',')
            text_file.write('\n')
```
