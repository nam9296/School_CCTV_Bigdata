```python
import pandas as pd
```

# 폭력지수 산출

### 취약점수 + 위해점수 - 경감점수 = 폭력지수

##### 1. 데이터 로드 및 전처리

```python
Vulnerability = pd.read_csv('data/Vulnerability_Indicator_final1.csv')
Hazard=pd.read_csv('data/hazard_indicator.csv')
Reduce = pd.read_csv('data/경감지표.csv',encoding='cp949')
```

- Vulnerability의 점수 칼럼 이름 수정

```python
Vulnerability = Vulnerability.rename({'score_2nd':'Vul_score'},axis=1)
```

```python
Vulnerability
```

- 위해지표의 지역칼럼을 지우고 점수 칼럼 이름 수정

```python
del Hazard['지역']
```

```python
Hazard = Hazard.rename({'score_2nd':'Haz_score'},axis=1)
```

```python
Hazard
```

- 위해지표와 취약지표 합치기

```python
Vulnerability_Hazard=pd.merge(Hazard,Vulnerability, how= 'inner', on='학교명')
Vulnerability_Hazard
```

```python
Vulnerability_Hazard.info()
```

- 취약점수 + 위해점수 

```python
Vulnerability_Hazard['Haz_Vul_score']= Vulnerability_Hazard['Haz_score']+Vulnerability_Hazard['Vul_score']
Vulnerability_Hazard
```

- 내림차순 점수 range 확인

```python
Vulnerability_Hazard.sort_values(by=['Haz_Vul_score'], ascending=[False])
```

- 경감지표에서 구 칼럼을 삭제하고 점수 칼럼 수정

```python
del Reduce['구']
```

```python
Reduce = Reduce.rename({'score_2nd':'Reduce_score'},axis=1)
```

```python
Reduce
```

- 경감점수를 기존 데이터와 merge

```python
Vulnerability_Hazard_Reduce=pd.merge(Vulnerability_Hazard,Reduce,how='inner',on='학교명')
```

- 취약점수와 위해점수의 합에 경감점수를 빼기

```python
Vulnerability_Hazard_Reduce['Final_Score'] = Vulnerability_Hazard_Reduce['Haz_Vul_score']-Vulnerability_Hazard_Reduce['Reduce_score']
Vulnerability_Hazard_Reduce
```

### 상위 20개 학교 확인

```python
top20 = Vulnerability_Hazard_Reduce.sort_values(by=['Final_Score'], ascending=[False]).head(20)
top20
```

```python
top20.reset_index(inplace=True, drop=True)
top20
```

```python
top30 = Vulnerability_Hazard_Reduce.sort_values(by=['Final_Score'], ascending=[False]).head(30)
top30
```

```python
top30.reset_index(inplace=True, drop=True)
top30
```

```python
Violence_Score20 = top20[['학교명','Final_Score','위도','경도']]
Violence_Score20
```

```python
Violence_Score20 = top20[['학교명','Final_Score','동','위도','경도']]
Violence_Score20
```

```python
Violence_Score30 = top30[['학교명','동','Final_Score','위도','경도']]
Violence_Score30
```

```python
Violence_Score20.to_csv('data/Violence_Score20.csv',index=False)
```
