## 구별 CCTV 대수 데이터

```python
CCTV = pd.read_csv('서울시 CCTV 설치운영 현황(자치구)-목적별.csv',encoding='cp949', thousands = ',')
CCTV
```

```python
CCTV = CCTV[['구분','방범','어린이\n보호구역']]
CCTV.head()
```

```python
CCTV = CCTV.rename({'구분':'구','방범':'방범용CCTV','어린이\n보호구역':'어린이보호구역CCTV'},axis=1)
CCTV
```

```python
CCTV.info()
```
