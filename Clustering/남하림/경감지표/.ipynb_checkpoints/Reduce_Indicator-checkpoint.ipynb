### 학교 500m 내 경찰관서수

```python
from tqdm.notebook import tqdm
from tqdm import tqdm, tqdm_notebook

from haversine import haversine
```

```python
def close_ps(school, Police_station, column) : 
    for i in tqdm(range(len(school))) : 
        distance = [] 
        for j in range(len(Police_station)) : 
            schoolloc = tuple(school[['경도','위도']].iloc[i])
            Police_stationloc = tuple(Police_station[['경도','위도']].iloc[j])
            distance.append(haversine(schoolloc, Police_stationloc, unit = 'm'))
            
        distance = [d for d in distance if d <= 500] 
        school[column][i] = len(distance)
        
    return school
```

```python
school["500미터내_파출소개수"] = 0
school = close_ps(school, Police_station, "500미터내_파출소개수")
```

## K-means Clustering

```python
school = pd.read_csv('./school_data/경감지수_전처리.csv')
school.info()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from scipy.spatial.distance import cdist, pdist

import matplotlib.pyplot as plt
```

```python
school_info = school[['학교코드','학교명','구','위도','경도']]
df = school[['내부상담_실시여부','외부상담_실시여부','wee클래스_운영여부','학교폭력_예방교육_평균교육시간(1학급당)','500미터내_파출소개수']]
labels= np.array(school['학교코드'])
```

```python
# 표준화
sd = StandardScaler()
df_sd = pd.DataFrame(sd.fit_transform(df))
df_sd.index = labels
```

```python
dist =[]
for i in range(1,20):
    k = KMeans(n_clusters=i,init='k-means++',
           n_init=10, max_iter=300, random_state=0)
    k.fit(df_sd)
    dist.append(k.inertia_)

plt.plot(range(1,20), dist, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.tight_layout()
plt.show()
```

```python
df_sd_arr = df_sd.values

#k값을 5로 설정 후 k-means에 적용시키기
kmeans = KMeans(n_clusters =5,random_state=2021).fit(df_sd_arr)
kmeans.cluster_centers_
```

```python
df_sd['cluster'] = kmeans.labels_
df_sd
```

```python
from sklearn.manifold import TSNE

model = TSNE(learning_rate=100)
transformed = model.fit_transform(df_sd.iloc[:,:5])

xs = transformed[:,0]
ys = transformed[:,1]
plt.scatter(xs,ys,c=kmeans.labels_)

plt.show()
```

```python
df_sd.reset_index(inplace=True)
df_sd.rename(columns={'index':'학교코드',0:'내부상담_실시여부',1:'외부상담_실시여부',2:'wee클래스_운영여부',3:'학교폭력_예방교육_평균교육시간(1학급당)',4:'500미터내_파출소개수'},inplace=True)
```

```python
df_sd['cluster'].value_counts()
```

```python
final = pd.merge(school_info, df_sd, on='학교코드', how='left')
final.head()
```

```python
final_score = final[['내부상담_실시여부','외부상담_실시여부','wee클래스_운영여부','학교폭력_예방교육_평균교육시간(1학급당)','500미터내_파출소개수','cluster']].groupby('cluster').mean()
final_score = final.groupby('cluster').mean()
final_score
```
