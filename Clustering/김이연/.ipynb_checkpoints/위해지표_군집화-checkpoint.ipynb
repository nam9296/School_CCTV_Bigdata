```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from scipy.spatial.distance import cdist, pdist

import matplotlib.pyplot as plt
```

```python
path = './school_data/'
```

```python
school = pd.read_csv(path+'위해지표_전처리_500m.csv')
school.head(10)
school.info()
```

```python
school['500미터내_청소년유해업소_계'] = school['500미터내_청소년출입금지업소개수']+school['500미터내_청소년고용금지업소개수']
school
```

```python
school_info = school[['학교코드','학교명','지역','법정동코드','위도','경도']]
df = school[['학교폭력유형_계','500미터내_청소년출입금지업소개수','500미터내_청소년고용금지업소개수','500미터내_청소년유해업소_계','개최횟수','심의건수','학교장의 자체해결건수','피해학생수','가해학생수','신체폭력','언어폭력','금품갈취','강요','따돌림','성폭력','사이버폭력','기타']]
labels= np.array(school['학교코드'])
```

```python
df.corr()
```

```python
# model 사용변수 선정

df = school[['학교폭력유형_계','500미터내_청소년출입금지업소개수','500미터내_청소년고용금지업소개수','개최횟수','심의건수','피해학생수','가해학생수','신체폭력','언어폭력']]
# df = school[['500미터내_청소년유해업소_계','개최횟수','심의건수','피해학생수','가해학생수','신체폭력','언어폭력']]
```

```python
# 표준화
sd = StandardScaler()
df_sd = pd.DataFrame(sd.fit_transform(df))
df_sd.index = labels
```

```python
dist =[]
for i in range(1,20):
    k = KMeans(n_clusters=i,init='k-means++',
           n_init=10, max_iter=300, random_state=0)
    k.fit(df_sd)
    dist.append(k.inertia_)

plt.plot(range(1,20), dist, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.tight_layout()
plt.show()
```

```python
df_sd_arr = df_sd.values

#k값을 5로 설정 후 k-means에 적용시키기
kmeans = KMeans(n_clusters = 5,random_state=2021).fit(df_sd_arr)
kmeans.cluster_centers_
```

```python
df_sd['cluster'] = kmeans.labels_
df_sd
```

```python
from sklearn.manifold import TSNE

model = TSNE(learning_rate=100)
transformed = model.fit_transform(df_sd.iloc[:,:7])

xs = transformed[:,0]
ys = transformed[:,1]
plt.scatter(xs,ys,c=kmeans.labels_)

plt.show()
```

```python
df_sd.reset_index(inplace=True)
df_sd.rename(columns={'index':'학교코드',0:'학교폭력유형_계',1:'500미터내_청소년출입금지업소개수',2:'500미터내_청소년고용금지업소개수',3:'개최횟수',4:'심의건수',5:'피해학생수',6:'가해학생수',7:'신체폭력',8:'언어폭력'},inplace=True)
```

```python
df_sd['cluster'].value_counts()
```

```python
final = pd.merge(school_info, df_sd, on='학교코드', how='left')
# final.head()
```

```python
final_score = final[['학교폭력유형_계','500미터내_청소년출입금지업소개수','500미터내_청소년고용금지업소개수','개최횟수','심의건수','피해학생수','가해학생수','신체폭력','언어폭력','cluster']].groupby('cluster').mean()
# final_score = final.groupby('cluster').mean()
final_score
```
