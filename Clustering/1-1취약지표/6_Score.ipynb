## 군집별 점수 부여

```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
score = pd.read_csv('data/Vulnerability_Cluster.csv')
score
```

```python
score['score'] = ''
```

```python
final_score = score[['학교명','학급수','학생수','학생수','학급당평균학생수','전출학생수','경제환경','교육환경','저소득','학업중단율','cluster']].groupby('cluster').mean()
final_score
```

```python
rank=final_score.sort_values('경제환경').index
rank
```

```python
# rank=final_score.sort_values('학급수').index
# rank
```

```python
for i in range(len(score)):
    if score['cluster'][i]==rank[0]:
        score['score'][i]=1
    elif score['cluster'][i]==rank[1]:
        score['score'][i]=0.7
    elif score['cluster'][i]==rank[2]:
        score['score'][i]=0.4
```

### 2차 점수 산출

- 1차 score에 선생1명당 학생수를 곱해주는 식으로 2차점수 측정

```python
data = pd.read_csv('data/new_data5.csv')
data
```

```python
new_data = data[['학교명','학생수','교원수']]
new_data
```

```python
new_data['선생1인당_학생수'] = new_data['학생수'] / new_data['교원수']
new_data
```

```python
score2 = pd.merge(score,new_data[['학교명','선생1인당_학생수']],how='inner',on='학교명')
score2

score2['score_2nd']=''

for i in range(len(score2)):
    for k in range(4):
        rank_lis = sorted(list(set(score2[score2['cluster']==k]['선생1인당_학생수'])), reverse=True) #한 군집내의 모든 인구밀도 내림차순 정렬
        for j in range(len(rank_lis)): #내림차순 정렬한 리스트의 인덱스 순서대로 적은 점수를 차감
            if score2['선생1인당_학생수'][i] == rank_lis[j]:
                score2['score_2nd'][i] = round((score2['score'][i]-(j/len(rank_lis))*0.2)*10,2) # 점수 만점을 10점으로 만들기

score2
```

```python
score2_final = score2[['학교명','자치구','score_2nd']]
score2_final_sorted=score2_final.sort_values(by=['score_2nd'], ascending=[False]).head(30)
```
