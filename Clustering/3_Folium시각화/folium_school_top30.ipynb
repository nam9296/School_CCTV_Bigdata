```python
import pandas as pd
import folium
import geopandas as gpd
```

```python
df = pd.read_csv('Violence_final_Score.csv')
df = df.sort_values(by = 'Final_Score', ascending=False).reset_index(drop=True)
df
```

```python
df_gu = df.groupby(['자치구'])['Final_Score'].mean().to_frame().reset_index()
df_gu
```

```python
#지도의 중심을 지정하기 위해 위도와 경도의 평균 구하기
lat = df['위도'].mean()
lon = df['경도'].mean()

print(lat)
print(lon)
```

```python
geo_data = gpd.read_file('서울시_행정구역경계.geojson')
geo_data
```

## 상위 30개 학교 시각화

```python
df_30 = df.iloc[:30]
df_30
```

```python
center = [37.541, 126.986]   # 서울시 중심

# 맵이 center 에 위치하고, zoom 레벨은 11로 시작하는 맵 m 생성
m = folium.Map(location=center, zoom_start=11,tiles = 'cartodb positron')

# Choropleth 레이어를 만들고, 맵 m에 추가
folium.Choropleth(
    geo_data=geo_data, data=df_final_gu, columns=('자치구', 'Final_Score'),
    key_on='feature.properties.name', fill_color='BuGn',
    legend_name='Final_Score',highlight =True).add_to(m)

for i in df_30.index:
    sub_lat = df_30.loc[i,'위도']
    sub_lon = df_30.loc[i,'경도']
    popup =df_30.loc[i,'학교명']
    title = df_30.loc[i,'Final_Score']
  
    # Final_Score이 11점이상이면 = red, default = 녹색
    color = 'blue'
    if df_30.loc[i,'Final_Score'] > 11:
        color = "red"

    folium.Marker([sub_lat,sub_lon], popup=popup,icon=folium.Icon(icon = 'bell', color=color)).add_to(m)

m
```

```python
m.save('school_top30.html')
```
