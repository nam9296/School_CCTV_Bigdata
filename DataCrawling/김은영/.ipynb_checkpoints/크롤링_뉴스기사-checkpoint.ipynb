## 반복해서 url 50 페이지 분량 가져오기

```python
url1 = 'https://search.hani.co.kr/Search?command=query&keyword='
url2='&media=news&submedia=&sort=s&period=all&datefrom=2012.01.01&dateto=2021.05.06&pageseq='
dic={} # 기사 담길 dict
index=0 # 기사번호 
keyword=str(input('검색할 키워드 입력 : '))
for i in tqdm(range(0,50)):
    url=url1+keyword+url2+str(i)
    req = requests.get(url)
    html = req.text
    soup = BeautifulSoup(html, 'html.parser')
    pages = soup.select('#contents > div.search-result-section > ul > li > dl > dt > a')
    for page in pages:
        index+=1
        tmp = page.get('href')
        dic[index]=tmp
```

```python
dic[1] # 1값당 1개의 url 들어가 있음을 확인할 수 있음.
```

## 텍스트 전처리 함수

```python
def preStr(txt):
    re.sub('[^A-Za-z0-9가-힣]', '', txt) 
    txt=txt.replace('\n','')
    return txt
```
